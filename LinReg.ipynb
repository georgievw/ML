```python
import numpy as np
import matplotlib.pyplot as plt

class MyLinearRegression():

  def mse(self, y_pred, y):
    return np.mean(np.square(y - y_pred))

  def fit(self, X_batch, y_batch, learning_rate=0.001, n_epochs=20):
    self.weights = np.zeros((X_batch.shape[1], 1))
    self.bias = float(0)
    for i in range(n_epochs):
      self.weights -= learning_rate*self.mse_gradient_w(self.predict(X_batch), y_batch, X_batch)
      self.bias -= learning_rate*self.mse_gradient_b(self.predict(X_batch), y_batch)

  def mse_gradient_w(self, y_pred, y, x):
    return -2 * np.mean((y - y_pred) * x)

  def mse_gradient_b(self, y_pred, y):
    return -2 * np.mean(y - y_pred)

  def predict(self, X_batch):
    return (X_batch.dot(self.weights) + self.bias)
```

```python
X = np.random.gumbel(50, 10, 1000).reshape(-1,1) 
y = X * 70000 
y = y + np.random.normal(0, 800000, (1000, 1)) 
```

```python
clf = MyLinearRegression()
clf.fit(X, y, 0.0001, 100)
plt.plot(X, y, 'o', alpha=0.3)
plt.plot(X, clf.predict(X))
plt.show()
```
