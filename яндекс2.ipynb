```python
import matplotlib.pyplot as plt
import numpy as np
import math
N = 1000
```

```python
#Окружность
M = 100
alpha = [2 * math.pi * i / M for i in range(M)]
xs_c = [math.cos(a) for a in alpha]
ys_c = [math.sin(a) for a in alpha]
```

```python
#Первый набор 
def generate1():
  abs = np.random.uniform(size=(N, 2))
  xs_1 = [ab[0] * math.cos(2 * math.pi * ab[1]) for ab in abs]
  ys_1 = [ab[0] * math.sin(2 * math.pi * ab[1]) for ab in abs]
  return [xs_1, ys_1]
```

```python
#Второй набор
def generate2_single():
    while True:  
        x = np.random.uniform(-1, 1)  
        y = np.random.uniform(-1, 1)  
        if x ** 2 + y ** 2 > 1:  
            continue  
        return [x, y]

def generate2():
  xy = np.array([generate2_single() for i in range(N)]).T
  xs_2, ys_2 = xy[0], xy[1]
  return [xs_2, ys_2]
```

```python
#Отрисовка всего
fig, axes = plt.subplots(nrows = 1, ncols = 2)
fig.set_figheight(5)
fig.set_figwidth(11)

xys1 = generate1()
axes[0].scatter(xys1[0], xys1[1]) 
axes[0].plot(xs_c, ys_c, color='red')

xys2 = generate2()
axes[1].scatter(xys2[0], xys2[1])
axes[1].plot(xs_c, ys_c, color='red')

plt.show()
```

```python
def r1_mean():
  xys1 = generate1()
  xs_1 = np.array(xys1[0])
  ys_1 = np.array(xys1[1])
  return np.sqrt(np.square(xs_1) + np.square(ys_1)).mean()
```

```python
def r2_mean():
  xys2 = generate2()
  xs_2 = np.array(xys2[0])
  ys_2 = np.array(xys2[1])
  return np.sqrt(np.square(xs_2) + np.square(ys_2)).mean()
```

```python
r1 = [r1_mean() for i in range(100)]
r2 = [r2_mean() for i in range(100)]
x = list(range(100))
fig, ax = plt.subplots()
fig.set_figheight(5)
fig.set_figwidth(5)

ax.plot(x, r1, color="red")
ax.plot(x, r2, color="green")
plt.show()
```
